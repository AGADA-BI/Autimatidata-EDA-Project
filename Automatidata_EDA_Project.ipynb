```python
# How many of each payment type are represented in the data?
"""" The output shows the frequency of different payment types in the DataFrame:

15,265 trips were paid with credit cards.
7,267 trips were paid with cash.
121 trips had no charge.
46 trips were disputed. """

df['payment_type'].value_counts()
```

```python
# Average tip for trips paid for with credit card.
average_cc_trip = df[df['payment_type'] == 1]['tip_amount'].mean()
print('Average credit cash trip:', average_cc_trip)

# Average tip for trips paid for with cash.
average_c_trip = df[df['payment_type']==2]['tip_amount'].mean()
print('Average cash trip:', average_c_trip)
```

```python
# Number of times each vendor ID is represented in the data. 
vendor_id_count = df['VendorID'].value_counts()
vendor_id_count
```

```python
# Mean total amount for each vendor.
df.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']]
```

```python
# Finding payments made with only Credit cards.
credit_card_payments = df[df['payment_type'] ==1]
# Using passenger counts to get the total number of passengers that paid with credit cards. This will in turn get the number of rides paid with credit cards.
'''From the output, we can see that a total of 10,977 payments were made with credit cards.'''
credit_card_payments['passenger_count'].value_counts()
```

```python
# Average tip amount for each passenger count (credit card payments only)
credit_card_payments.groupby(['passenger_count']).mean(numeric_only = True)[['tip_amount']]
```
